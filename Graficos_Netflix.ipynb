```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('netflix.csv')
```

```python
df = df.drop(df.columns[0], axis=1)
```

```python
df.head()
```

```python
eixo_x = df['Score'].value_counts().keys()
eixo_y = df['Score'].value_counts().values
```

```python
#Configurar gráfico de barras com eixos X e Y

plt.bar(eixo_x, eixo_y)

#Adicionar título do gráfico

plt.title('Ranking de Filmes')

#Adicionar rotação aos dados do eixo X do gráfico (neste caso em 45 graus)

plt.xticks(rotation=45)

#Adicionar títulos aos eixos X e Y

plt.xlabel('Nota do Ranking')
plt.ylabel('Numero de filmes')

#Mostrar gráfico
plt.show
```

**Gráfico Ranking x Ano de Lançamento**

```python
eixo_x = df['Year'].value_counts().sort_index().keys()
eixo_y = df['Year'].value_counts().sort_index().values
```

```python
#Configurar gráfico de barras com eixos X e Y

plt.plot(eixo_x, eixo_y)

#Adicionar título do gráfico

plt.title('Filmes produzidos por ano')

#Adicionar rotação aos dados do eixo X do gráfico (neste caso em 45 graus)

plt.xticks(rotation=45)

#Adicionar títulos aos eixos X e Y

plt.xlabel('Ano do Filme')
plt.ylabel('Numero de filmes')

#Mostrar gráfico
plt.show
```

```python
#Configurar gráfico de barras com eixos X e Y

plt.scatter(eixo_x, eixo_y)

#Adicionar título do gráfico

plt.title('Filmes produzidos por ano')

#Adicionar rotação aos dados do eixo X do gráfico (neste caso em 45 graus)

plt.xticks(rotation=45)

#Adicionar títulos aos eixos X e Y

plt.xlabel('Ano do Filme')
plt.ylabel('Numero de filmes')

#Mostrar gráfico
plt.show
```

```python
# Para identificar os diretores que mais produziram filmes

df['Director'].value_counts().head(10).plot(kind='barh')
plt.title('Quantidade de Filmes por Diretor')
plt.xlabel('Número de Filmes')
plt.ylabel('Diretor')
plt.show()
```

```python
# Ranking de notas de cada diretor

top_directors = df.groupby('Director')['Score'].mean().sort_values(ascending=False).head(10)
top_directors.plot(kind='barh')
plt.title('Ranking de Notas por Diretor')
plt.xlabel('Média da Nota')
plt.ylabel('Diretor')
plt.show()
```

```python
# Notas ao Longo dos Anos

df.groupby('Year')['Score'].mean().plot(kind='line', marker='o')
plt.title('Notas ao Longo dos Anos')
plt.xlabel('Ano')
plt.ylabel('Nota Média')
plt.grid()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Carregar os dados
df = pd.read_csv('netflix.csv')

# Remover a primeira coluna (se for um índice desnecessário)
df = df.drop(df.columns[0], axis=1)

# Ordenar os filmes pelos mais bem avaliados
top_filmes = df.sort_values(by='Score', ascending=False).head(10)

# Criar uma nova coluna combinando Título e Diretor
top_filmes['Filme_Diretor'] = top_filmes['Movie Title'] + '\n(' + top_filmes['Director'] + ')'

# Criar gráfico de barras
plt.figure(figsize=(10, 6))
plt.barh(top_filmes['Filme_Diretor'], top_filmes['Score'])
plt.xlabel('Nota')
plt.ylabel('Filme (Diretor)')
plt.title('Top 10 Filmes Mais Bem Avaliados')
plt.gca().invert_yaxis()  # Inverter para mostrar o melhor no topo

# Mostrar gráfico
plt.show()

```
